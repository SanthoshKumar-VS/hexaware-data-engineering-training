```python
from pyspark.sql.window import Window
from pyspark.sql.functions import rank, row_number
window_spec = Window.partitionBy("department").orderBy(df_emp.salary.desc())
df_emp.withColumn("rank", rank().over(window_spec)).show()
```

```python
from pyspark.sql.functions import col, sum
df_sales = df_sales.withColumn("amount", col("amount").cast("int"))
df_sales.select(sum("amount")).show()
df_sales.groupBy("emp_id").sum("amount").show()
df_sales.groupBy("emp_id") \
        .sum("amount") \
        .orderBy("sum(amount)", ascending=False) \
        .show(1)
df_sales.groupBy("product").sum("amount").show()
```

```python
from pyspark.sql.functions import udf
from pyspark.sql.types import StringType

def categorize(amount):
    if amount > 50000:
        return "Premium"
    elif amount >= 20000:
        return "Medium"
    else:
        return "Small"

cat_udf = udf(categorize, StringType())
df_sales.withColumn("sale_category", cat_udf(df_sales.amount)).show()
```
