**Section 5 — Conditional Queries**

```sql
select * from employees where salary > (select avg(salary) from employees);
```

```sql
select * from employees e 
where experience > (select avg(experience) 
from employees 
where department = e.department);
```

```sql
select * from employees where salary < 70000;
```

```sql
select * from employees where department = 'HR';
```

```sql
select * from employees where department = 'Finance' and salary > 90000;
```

**Section 6 — Joins**

```sql
select e.emp_name, e.department, p.budget
from employees e
join projects p
on e.department = p.department;
```

```sql
select e.emp_name, e.department, p.project_name
from employees e
join projects p
on e.department = p.department;
```

```sql
select p.department, sum(p.budget)
from projects p
group by p.department;
```

```sql
select distinct e.*
from employees e
join projects p
on e.department = p.department;
```

```sql
select p.project_name, count(e.emp_id)
from employees e
join projects p
on e.department = p.department
group by p.project_name;
```

**Section 7 — Advanced Queries**

```sql
select * from employees
where department = 'Engineering' order by salary desc limit 3;
```

```sql
select department from employees
group by department order by avg(salary) desc limit 1;
```

```sql
select city from employees 
group by city order by count(*) desc limit 1;
```

```sql
select * from employees e 
where salary > (select avg(salary) 
from employees where department = e.department);
```

```sql
select * from employees e 
where experience = (select max(experience)
from employees where department = e.department);
```

**Final Mini Capstone**

```sql
select emp_name, city, department, salary, experience from employees;
```

```sql
select emp_name, city, department, salary, experience,
case
when salary >= 95000 then 'High'
when salary >= 80000 then 'Medium'
else 'Low'
end as salary_category
from employees;
```

```sql
select emp_name, city, department, salary, experience,
case
when salary >= 95000 then 'High'
when salary >= 80000 then 'Medium'
else 'Low'
end as salary_category,
avg(salary) over(partition by department) as department_avg_salary
from employees;
```

```sql
create table employee_salary_analysis as
select emp_name, city, department, salary, experience,
case
when salary >= 95000 then 'High'
when salary >= 80000 then 'Medium'
else 'Low'
end as salary_category,
avg(salary) over(partition by department) as department_avg_salary
from employees;
```
