Final Mini Project (RDD)

1.Total revenue per product

```python
total_revenue_per_product = sales_rdd.map(lambda x: (x[0], x[1]*x[2])) \
.reduceByKey(lambda a,b: a+b)
total_revenue_per_product.collect()
```

2.Top selling product

```python
top_selling_product = sales_rdd.map(lambda x: (x[0], x[1])) \
.reduceByKey(lambda a,b: a+b) \
.sortBy(lambda x: x[1], ascending=False) \
.take(1)
top_selling_product
```

4. Products sold more than 3 times
