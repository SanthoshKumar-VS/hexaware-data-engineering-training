```python
from pyspark.sql import SparkSession

spark = SparkSession.builder\
.appName("PySpark Exercises")\
.getOrCreate()
```

```python
data = [
    ("Santhosh Kumar", 33000, "Data Engineer"),
    ("Suresh Krishna", 35000, "Dotnet Developer"),
    ("Jason", 30000, "Tester"),
    ("Yaadhav", 52000, "Java Developer"),
    ("Manohar", 30000, "Tester")
]

columns = ["Name", "Salary", "Department"]

df = spark.createDataFrame(data, columns)

df.show()
```

```python
df.select("Name", "Salary").show()
```

```python
df.filter(df.Salary > 50000).show()
```

```python
df.groupBy("Department").count().show()
```

```python
df.orderBy("Salary").show()
```
