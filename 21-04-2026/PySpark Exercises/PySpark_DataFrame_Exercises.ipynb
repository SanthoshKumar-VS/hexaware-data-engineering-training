```python
df_emp = spark.read.csv("employees.csv", header=True, inferSchema=True)
df_emp.show()
```

```python
df_dept = spark.read.csv("departments.csv", header=True, inferSchema=True)
df_dept.show()
```

```python
df_sales = spark.read.csv("sales.csv", header=True, inferSchema=True)
df_sales.show()
```

```python
from pyspark.sql.functions import col
df_emp.filter(col("joining_date") > "2023-01-01").show()
```

```python
df_emp.filter(df_emp.salary > 65000).show()
```

```python
df_emp.filter(df_emp.department == "IT").show()
```

```python
df_emp.filter((df_emp.department == "IT") & (df_emp.salary > 70000)).show()
```

```python
df_emp.drop("joining_date").show()
```

```python
df_emp.orderBy(df_emp.salary.asc()).show()
```

```python
df_emp.orderBy(df_emp.salary.desc()).show()
```

```python
df_emp.orderBy(df_emp.salary.desc()).show(3)
```

```python
from pyspark.sql.functions import sum, avg, max, min
df_emp.select(sum("salary")).show()
```

```python
df_emp.select(avg("salary")).show()
```

```python
df_emp.select(max("salary")).show()
```

```python
df_emp.select(min("salary")).show()
```

```python
df_emp.groupBy("department").count().show()
```

```python
df_emp.join(df_dept, on="department", how="inner").show()
```

```python
df_emp.join(df_dept, on="department", how="left").show()
```

```python
df_emp.join(df_dept, on="department")\
.select("name", "location").show()
```

```python
from pyspark.sql.functions import lit
df_emp.withColumn("bonus", df_emp.salary * 0.1).show()
```

```python
df_emp.withColumn("company", lit("BotCampus")).show()
```

```python
df_emp.filter(df_emp.department.isNull()).show()
```

```python
df_emp.fillna({"department": "Unknown"}).show()
```

```python
df_emp.dropna().show()
```

```python
from pyspark.sql.functions import upper, length
df_emp.select(
    "name",
    upper("name").alias("upper_name")
).show()
```

```python
df_emp.select(
    length("name").alias("name_length")
).show()
```

```python
df_emp.withColumn("salary_in_lakhs", df_emp.salary / 100000).show()
```
