**Part 2 — DataFrame Transformations**

```python
from pyspark.sql.functions import *
patients_df = patients_df.withColumn(
    "registration_date",
    to_date(col("registration_date"), "yyyy-MM-dd")
)
patients_df.show()
```

```python
visits_df = visits_df.withColumn(
    "visit_date",
    to_date(col("visit_date"), "yyyy-MM-dd")
)

visits_df.show()
```

```python
visits_df = visits_df.withColumn(
    "test_cost",
    col("tests_count") * 1500
)
visits_df.show()
```

```python
visits_df = visits_df.withColumn(
    "estimated_total_bill",
    col("test_cost") + 2000
)

visits_df.show()
```

```python
patients_df = patients_df.withColumn(
    "age_category",
    when(col("age") < 30, "Young")
    .when(col("age").between(30, 45), "Adult")
    .otherwise("Senior")
)

patients_df.show()
```

```python
patients_df = patients_df.withColumn(
    "membership_priority",
    when(col("membership") == "Gold", "High")
    .when(col("membership") == "Silver", "Medium")
    .otherwise("Low")
)

patients_df.show()
```

```python
visits_df = visits_df.withColumn(
    "visit_priority",
    when(col("tests_count") >= 3, "High")
    .otherwise("Normal")
)

visits_df.show()
```

```python
payments_df = payments_df.withColumn(
    "high_bill_flag",
    when(col("bill_amount") > 7000, "Yes")
    .otherwise("No")
)

payments_df.show()
```

```python
patients_df = patients_df.withColumnRenamed(
    "patient_name",
    "full_name"
)

patients_df.show()
```

```python
payments_df = payments_df.drop("high_bill_flag")

payments_df.show()
```

**Part 6 — Window Functions**

```python
from pyspark.sql.window import Window

city_window = Window.partitionBy("city")\
.orderBy(col("bill_amount").desc())

final_joined_df.withColumn("rank",
rank().over(city_window)).select(
"full_name","city","bill_amount","rank"
).show()
```

```python
spec_window = Window.partitionBy("specialization")\
.orderBy(col("bill_amount").desc())

final_joined_df.withColumn("rank",
rank().over(spec_window)).select(
"doctor_name","specialization",
"bill_amount","rank"
).show()
```

```python
row_window = Window.partitionBy("city")\
.orderBy(col("bill_amount").desc())

final_joined_df.withColumn("row_num",
row_number().over(row_window))\
.filter(col("row_num")==1)\
.select("full_name","city","bill_amount").show()
```

```python
dense_window = Window.partitionBy("specialization")\
.orderBy(col("consultation_fee").desc())

final_joined_df.withColumn("dense_rank",
dense_rank().over(dense_window)).select(
"doctor_name","specialization",
"consultation_fee","dense_rank"
).show()
```

```python
doctor_revenue = final_joined_df.groupBy(
"doctor_name").agg(
sum("bill_amount").alias("total_revenue")
)

top_window = Window.orderBy(
col("total_revenue").desc()
)

doctor_revenue.withColumn("rank",
rank().over(top_window))\
.filter(col("rank")<=2).show()
```

```python
date_window = Window.orderBy("visit_date")

final_joined_df.withColumn("running_total",
sum("bill_amount").over(date_window))\
.select("visit_date","bill_amount",
"running_total").show()
```

```python
doctor_window = Window.partitionBy(
"doctor_name").orderBy("visit_date")

final_joined_df.withColumn("running_total",
sum("bill_amount").over(doctor_window))\
.select("doctor_name","visit_date",
"bill_amount","running_total").show()
```

```python
city_revenue = final_joined_df.groupBy(
"city").agg(
sum("bill_amount").alias("total_revenue")
)

city_window = Window.orderBy(
col("total_revenue").desc()
)

city_revenue.withColumn("rank",
rank().over(city_window)).show()
```

```python
spec_revenue = final_joined_df.groupBy(
"specialization").agg(
sum("bill_amount").alias("total_revenue")
)

spec_window = Window.orderBy(
col("total_revenue").desc()
)

spec_revenue.withColumn("rank",
rank().over(spec_window)).show()
```

```python
payment_window = Window.partitionBy(
"payment_mode").orderBy(
col("bill_amount").desc()
)

final_joined_df.withColumn("rank",
row_number().over(payment_window))\
.filter(col("rank")==1)\
.select("payment_mode",
"bill_amount","full_name").show()
```

```python
daily_visits_data = [
(26,1002,201,"2024-03-14","Completed",2),
(27,1005,210,"2024-03-14","Completed",1),
(11,1011,205,"2024-03-06","Completed",3),
(20,1020,206,"2024-03-10","Completed",2),
(28,1018,203,"2024-03-15","Pending",2)
]

daily_visits_columns = [
"visit_id","patient_id","doctor_id","visit_date","visit_status","tests_count"
]

daily_visits_df = spark.createDataFrame(daily_visits_data, daily_visits_columns)
```

```python
visits_df.write.format("delta")\
.mode("overwrite")\
.save("patient_delta")
```

```python
target_df = spark.read.format("delta")\
.load("patient_delta")

target_df.show()
```

```python
daily_visits_df.createOrReplaceTempView(
"new_data"
)
```

```python
from delta.tables import DeltaTable

delta_table = DeltaTable.forPath(
spark,"patient_delta"
)

delta_table.alias("target").merge(
daily_visits_df.alias("source"),
"target.visit_id = source.visit_id"
).whenMatchedUpdate(set={
"patient_id":"source.patient_id",
"doctor_id":"source.doctor_id",
"visit_date":"source.visit_date",
"visit_status":"source.visit_status",
"tests_count":"source.tests_count"
}).whenNotMatchedInsert(values={
"visit_id":"source.visit_id",
"patient_id":"source.patient_id",
"doctor_id":"source.doctor_id",
"visit_date":"source.visit_date",
"visit_status":"source.visit_status",
"tests_count":"source.tests_count"
}).execute()
```

**Part 10 — DLT Assessment**

```python
import dlt
from pyspark.sql.functions import *
```

```python
@dlt.table
def bronze_patient_visits():
    return final_joined_df
```

```python
@dlt.table
def silver_patient_visits():

    return dlt.read("bronze_patient_visits")\
    .withColumn(
    "total_bill",
    col("bill_amount")+col("consultation_fee")
    ).filter(
    col("visit_status")!="Cancelled"
    )
```

```python
@dlt.table
def gold_city_revenue():

    return dlt.read("silver_patient_visits")\
    .groupBy("city")\
    .agg(
    sum("total_bill").alias("city_revenue")
    )
```

```python
@dlt.table
def gold_specialization_revenue():

    return dlt.read("silver_patient_visits")\
    .groupBy("specialization")\
    .agg(
    sum("total_bill")\
    .alias("specialization_revenue")
    )
```
