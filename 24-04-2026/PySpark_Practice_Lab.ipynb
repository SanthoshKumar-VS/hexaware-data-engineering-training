```python
from pyspark.sql.functions import sum
df = orders.join(customers, "customer_id")\
.join(order_items, "order_id")\
.join(products, "product_id")\
.withColumn("revenue", order_items.quantity * products.price)\
.groupBy("name")\
.agg(sum("revenue").alias("total_revenue"))
w = Window.orderBy(df.total_revenue.desc())
df.withColumn("rank", rank().over(w)).show()
```

```python
from pyspark.sql.functions import dense_rank
w = Window.orderBy(df.total_revenue.desc())
df.withColumn("dense_rank", dense_rank().over(w)).show()
```

```python
from pyspark.sql.functions import col,sum,rank
from pyspark.sql.window import Window
df=customers.join(orders,"customer_id")\
.join(order_items,"order_id")\
.join(products,"product_id")\
.withColumn("revenue",col("quantity")*col("price"))
df2=df.groupBy("name","city")\
.agg(sum("revenue").alias("total_revenue"))
w=Window.partitionBy("city").orderBy(col("total_revenue").desc())
df2.withColumn("rank",rank().over(w))\
.filter("rank=1")\
.show()
```

```python
df_prod = order_items.join(products, "product_id")\
.withColumn("revenue", order_items.quantity * products.price)\
.groupBy("product_name", "category")\
.agg(sum("revenue").alias("total_revenue"))
w = Window.partitionBy("category").orderBy(df_prod.total_revenue.desc())
df_prod.withColumn("rank", rank().over(w))\
.filter("rank = 1")\
.show()
```

```python
spark.sql("""
SELECT c.name, SUM(oi.quantity * p.price) AS revenue
FROM customers c
JOIN orders o ON c.customer_id = o.customer_id
JOIN order_items oi ON o.order_id = oi.order_id
JOIN products p ON oi.product_id = p.product_id
GROUP BY c.name
ORDER BY revenue DESC
LIMIT 3
""").show()
```

```python
spark.sql("""
SELECT p.product_name, SUM(oi.quantity * p.price) AS revenue
FROM order_items oi
JOIN products p ON oi.product_id = p.product_id
GROUP BY p.product_name
ORDER BY revenue DESC
LIMIT 1
""").show()
```

Final Capstone Mini Project

Join all tables

```python
df = customers.join(orders, "customer_id")\
.join(order_items, "order_id")\
.join(products, "product_id")\
.join(payments, "order_id")\
.withColumn("revenue", col("quantity") * col("price"))
df.show()
```

Calculate revenue

```python
from pyspark.sql.functions import col
df = orders.join(order_items, "order_id")\
.join(products, "product_id")\
.withColumn("revenue", col("quantity") * col("price"))
df.select("order_id", "product_id", "quantity", "price", "revenue").show()
```

top customers

```python
from pyspark.sql.functions import sum
top_customers = df.groupBy("name")\
.agg(sum("revenue").alias("total_revenue"))\
.orderBy(col("total_revenue").desc())
top_customers.show()
```

top products

```python
top_products = df.groupBy("product_name")\
.agg(sum("revenue").alias("total_revenue"))\
.orderBy(col("total_revenue").desc())
top_products.show()
```

revenue per city

```python
from pyspark.sql.functions import col, sum
df = customers.join(orders, "customer_id")\
.join(order_items, "order_id")\
.join(products, "product_id")\
.withColumn("revenue", col("quantity") * col("price"))
revenue_city = df.groupBy("city")\
.agg(sum("revenue").alias("total_revenue"))
revenue_city.show()
```

Rank customers by revenue

```python
from pyspark.sql.window import Window
from pyspark.sql.functions import rank
w = Window.orderBy(col("total_revenue").desc())
ranked_customers = top_customers.withColumn("rank", rank().over(w))
ranked_customers.show()
```
