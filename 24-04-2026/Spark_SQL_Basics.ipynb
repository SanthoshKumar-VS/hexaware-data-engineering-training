Filtering in SQL means selecting only rows that satisfy a condition using WHERE.

```python
spark.sql("""
SELECT name, city
FROM employees
WHERE department = 'IT'
""").show()
```

```python
spark.sql("""
SELECT *
FROM employees
WHERE city IN ('Hyderabad', 'Bangalore')
""").show()
```

JOIN is used to combine data from two tables based on a common column.

```python
spark.sql("""
SELECT e.emp_id, e.name, e.department, d.dept_full_name
FROM employees e
INNER JOIN departments d
ON e.department = d.department
""").show()
```

LEFT JOIN returns all records from left table and matching records from right table.

```python
spark.sql("""
SELECT e.emp_id, e.name, e.department, d.dept_full_name
FROM employees e
LEFT JOIN departments d
ON e.department = d.department
""").show()
```

RIGHT JOIN returns all records from the right table and matching records from the left table.

```python
spark.sql("""
SELECT e.emp_id, e.name, e.department, d.dept_full_name
FROM employees e
RIGHT JOIN departments d
ON e.department = d.department
""").show()
```

Aggregation is used to perform calculations like count, sum, avg, max, min on data.

```python
spark.sql("""
SELECT
COUNT(*) AS total_employees,
SUM(salary) AS total_salary,
AVG(salary) AS avg_salary,
MAX(salary) AS max_salary,
MIN(salary) AS min_salary
FROM employees
""").show()
```

GROUP BY is used to group rows based on a column and apply aggregation functions.

```python
spark.sql("""
SELECT department, COUNT(*) AS emp_count
FROM employees
GROUP BY department
""").show()
```

AVG() calculates the average value of a column.

```python
spark.sql("""
SELECT department, AVG(salary) AS avg_salary
FROM employees
GROUP BY department
""").show()
```

Combine tables using JOIN and calculate aggregated values.

```python
spark.sql("""
SELECT e.name, SUM(s.quantity * s.price) AS total_sales
FROM employees e
JOIN sales s
ON e.emp_id = s.emp_id
GROUP BY e.name
""").show()
```

HAVING is used to filter grouped data after applying aggregation.

```python
spark.sql("""
SELECT department, AVG(salary) AS avg_salary
FROM employees
GROUP BY department
HAVING AVG(salary) > 65000
""").show()
```

Use HAVING with aggregation like SUM() to filter grouped results.

```python
spark.sql("""
SELECT product, SUM(quantity) AS total_qty
FROM sales
GROUP BY product
HAVING SUM(quantity) > 2
""").show()
```

ORDER BY is used to sort the result based on a column (ascending or descending).

```python
spark.sql("""
SELECT *
FROM employees
ORDER BY salary DESC
""").show()
```

Use ORDER BY with aggregated columns to sort calculated results.

```python
spark.sql("""
SELECT product, SUM(quantity * price) AS revenue
FROM sales
GROUP BY product
ORDER BY revenue DESC
""").show()
```

Window functions perform calculations across a group of rows without grouping them into a single row.

RANK() assigns ranking to rows within each group based on a specified order.

```python
spark.sql("""
SELECT
    emp_id,
    name,
    department,
    salary,
    RANK() OVER (PARTITION BY department ORDER BY salary DESC) AS salary_rank
FROM employees
""").show()
```

ROW_NUMBER() assigns a unique sequential number to each row within a partition.

```python
spark.sql("""
SELECT
    emp_id,
    name,
    department,
    salary,
    ROW_NUMBER() OVER (PARTITION BY department ORDER BY salary DESC) AS row_num
FROM employees
""").show()
```

DENSE_RANK() assigns ranks without gaps between ranking values within each group.

```python
spark.sql("""
SELECT
    emp_id,
    name,
    department,
    salary,
    DENSE_RANK() OVER (PARTITION BY department ORDER BY salary DESC) AS dense_rank
FROM employees
""").show()
```
